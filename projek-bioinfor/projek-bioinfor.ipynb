```python
# Read genbank file
organisme_basmah_forest = "sequence.gb"
```

```python
# Import biopython modules
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
for record in SeqIO.parse(organisme_basmah_forest, "genbank"):
    print(record.id)
    print(record.description)
    print(record.seq)
    print(len(record.seq))
```

# No 1

```python

# tampilkan jumlah setiap karakter basa yang ada di data tanpa terkecuali
print("Jumlah setiap karakter basa yang ada di data tanpa terkecuali")
print("Jumlah Basa Adenine :",record.seq.count("A"))
print("Jumlah Basa Thmine :", record.seq.count("T"))
print("Jumlah Basa Guanine : ", record.seq.count("G"))
print("Jumlah Basa Cytosine: ", record.seq.count("C"))
print("Jumalah Karakter Basa Lainnya: ", record.seq.count("N"))

```

```python
""" 
Dan pada data yang didapatkan, CDS pada gen yang berisikan 
urutan asam amino juga akan Anda hadirkan secara kumulatif. Anda 
menggabungkan semua CDS dalam satu urutan dan memberikan laporan jumlah
"""

tag = 0
# membuat variabel kosong untuk melakukan handling file

# Hadirkan cds pada gen yang berisikan urutan asam amino secara kumulatif
print("CDS pada gen yang berisikan urutan asam amino secara kumulatif")
for feature in record.features:
    if feature.type == "CDS":
        # Use try and except to avoid errors if the feature does not have a translation
        try:
            print(feature.qualifiers["translation"][0])
            with open("cds.txt", "a") as f:
                f.write(feature.qualifiers["translation"][0])
                f.write("\n")

            tag += 1
        except:
            pass
           

# Hadirkan jumlah setiap karakter basa yang ada di data tanpa terkecuali

```

```python
# Asign cds to variable
cds = record.features[0]
print(cds)
print(cds.type)

structural_polyprotein = []

# Find cds['structural polyprotein']
for i in range(len(record.features)):
    if record.features[i].type == "CDS" and record.features[i].qualifiers['product'][0] == "structural polyprotein":
        structural_polyprotein.append(record.features[i])


# show location of cds['structural polyprotein']
print(structural_polyprotein[0].location)

# show sequence of cds['structural polyprotein']



structural_polyprotein = structural_polyprotein[0]

# Print structural_polyprotein
print(structural_polyprotein)

# save structural_polyprotein_seq to file
with open("structural_polyprotein_seq.txt", "w") as f:
    f.write(str(structural_polyprotein.qualifiers['translation']))
    




```
